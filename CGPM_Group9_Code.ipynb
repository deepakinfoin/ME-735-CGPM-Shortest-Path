```python
#n=int(input("Enter size of the pixel grid "))
n=20
mat= np.zeros((n,n))

B=np.ones((9,4))
B_ipoints=[0,0]
B_fpoints=B_ipoints+np.array(B.shape)
mat[B_ipoints[0]:B_fpoints[0],B_ipoints[1]:B_fpoints[1]]=B

C=np.ones((5,3))
C_ipoints=[0,5]
C_fpoints=C_ipoints+np.array(C.shape)
mat[C_ipoints[0]:C_fpoints[0],C_ipoints[1]:C_fpoints[1]]=C

D=np.ones((3,6))
D_ipoints=[0,9]
D_fpoints=D_ipoints+np.array(D.shape)
mat[D_ipoints[0]:D_fpoints[0],D_ipoints[1]:D_fpoints[1]]=D

E=np.ones((4,4))
E_ipoints=[0,16]
E_fpoints=E_ipoints+np.array(E.shape)
mat[E_ipoints[0]:E_fpoints[0],E_ipoints[1]:E_fpoints[1]]=E

F=np.ones((4,3))
F_ipoints=[6,5]
F_fpoints=F_ipoints+np.array(F.shape)
mat[F_ipoints[0]:F_fpoints[0],F_ipoints[1]:F_fpoints[1]]=F

G=np.ones((4,4))
G_ipoints=[4,9]
G_fpoints=G_ipoints+np.array(G.shape)
mat[G_ipoints[0]:G_fpoints[0],G_ipoints[1]:G_fpoints[1]]=G

H=np.ones((4,4))
H_ipoints=[5,16]
H_fpoints=H_ipoints+np.array(H.shape)
mat[H_ipoints[0]:H_fpoints[0],H_ipoints[1]:H_fpoints[1]]=H

I=np.ones((4,4))
I_ipoints=[10,0]
I_fpoints=I_ipoints+np.array(I.shape)
mat[I_ipoints[0]:I_fpoints[0],I_ipoints[1]:I_fpoints[1]]=I

J=np.ones((4,3))
J_ipoints=[10,5]
J_fpoints=J_ipoints+np.array(J.shape)
mat[J_ipoints[0]:J_fpoints[0],J_ipoints[1]:J_fpoints[1]]=J

K=np.ones((2,6))
K_ipoints=[9,9]
K_fpoints=K_ipoints+np.array(K.shape)
mat[K_ipoints[0]:K_fpoints[0],K_ipoints[1]:K_fpoints[1]]=K

K1=np.ones((2,6))
K1_ipoints=[12,9]
K1_fpoints=K1_ipoints+np.array(K1.shape)
mat[K1_ipoints[0]:K1_fpoints[0],K1_ipoints[1]:K1_fpoints[1]]=K1

L=np.ones((3,4))
L_ipoints=[10,16]
L_fpoints=L_ipoints+np.array(L.shape)
mat[L_ipoints[0]:L_fpoints[0],L_ipoints[1]:L_fpoints[1]]=L

M=np.ones((5,6))
M_ipoints=[15,0]
M_fpoints=M_ipoints+np.array(M.shape)
mat[M_ipoints[0]:M_fpoints[0],M_ipoints[1]:M_fpoints[1]]=M

N=np.ones((5,4))
N_ipoints=[15,7]
N_fpoints=N_ipoints+np.array(N.shape)
mat[N_ipoints[0]:N_fpoints[0],N_ipoints[1]:N_fpoints[1]]=N

O=np.ones((5,3))
O_ipoints=[15,12]
O_fpoints=O_ipoints+np.array(O.shape)
mat[O_ipoints[0]:O_fpoints[0],O_ipoints[1]:O_fpoints[1]]=O

P=np.ones((6,4))
P_ipoints=[14,16]
P_fpoints=P_ipoints+np.array(P.shape)
mat[P_ipoints[0]:P_fpoints[0],P_ipoints[1]:P_fpoints[1]]=P

Q=np.ones((4,1))
Q_ipoints=[4,14]
Q_fpoints=Q_ipoints+np.array(Q.shape)
mat[Q_ipoints[0]:Q_fpoints[0],Q_ipoints[1]:Q_fpoints[1]]=Q

imat=mat
mat=np.ones((n,n))-imat    #For inverting colors
print(mat)

plt.matshow(mat)
plt.show()
```

```python
node=[]
o_node=[]
i_node=[]
for j in range(n):                 #For Outer Nodes
    if mat[0][j]==1:
        m=[0,j]
        o_node.append(m)
        
    if mat[j][0]==1:
        m=[j,0]
        o_node.append(m)
        
    if mat[n-1][j]==1:
        m=[n-1,j]
        o_node.append(m)
        
    if mat[j][n-1]==1:
        m=[j,n-1]
        o_node.append(m)

        
for i in range(1,n-1):            #For Inner nodes
    for j in range(1,n-1):
        sum=mat[i][j]+mat[i-1][j]+mat[i+1][j]+mat[i][j-1]+mat[i][j+1]
        if sum>=4:
            m=[i,j]
            i_node.append(m)
           
        
node=o_node+i_node

l=len(node)

node=sorted(node, key=lambda i:(i[0],i[1]))   #Rearranging nodes in ascending order
print(node)
    

#Initially distance between all the nodes are kept 0, so they are not connected.
distance=np.zeros((l,l))

def summ(i,k,j):
    step=int((k-i)/abs(k-i))     #To make it 1 or -1
    jd=1-j                       #jd is 1 if j=0 and 0 if j=1

    el=[m for m in range(node[i][jd],node[k][jd],step)] #Extract rows position if jd=0, columns position if jd=1
    if j==0:
        r=node[i][j]
        elem=mat[r,el]
    else:
        c=node[i][j]
        elem=mat[el,c]
    return np.sum(elem)
    

for i in range(l):                 #Iterate over each node
    for j in range(2):             #Iterate over row and column
        for k in range(l):         #Iterate over each node
            if node[i][j]==node[k][j] and (node[i] not in o_node or node[k] not in o_node):
                    A=node[i]
                    B=node[k]
                    distance[i][k]=max(abs(np.array(A)-np.array(B)))
                    if i!=k and (distance[i][k]>summ(i,k,j)):  #For removing data including distances between 2 nodes over the obstacles
                        distance[i][k]=0
   

print(distance)  #Adjacency List
```

```python
matrix=distance
```

```python
def MatrixtoDict(m):         #defining a function to convert adjacency list to Dictionary list

    graph = {}
    for idx, row in enumerate(m):
        res = {}
        for r in range(len(row)):
            if row[r] != 0:
                res[r]=row[r]
            graph[idx] = res
    return graph

mat_dict=MatrixtoDict(matrix)
print(mat_dict)
```

```python
#code implemented from stackoverflow https://stackoverflow.com/questions/22897209/dijkstras-algorithm-in-python/61078380#61078380
class Dijkstra:

    def __init__(self, vertices, graph):
        self.vertices = vertices  # ("A", "B", "C" ...)
        self.graph = graph  # {"A": {"B": 1}, "B": {"A": 1, "C": 5} ...}

    def find_route(self, start, end):
        unvisited = {n: float("inf") for n in self.vertices}
        unvisited[start] = 0  # set start vertex to 0
        visited = {}  # list of all visited nodes
        parents = {}  # predecessors
        while unvisited:
            min_vertex = min(unvisited, key=unvisited.get)  # get smallest distance
            for neighbour, _ in self.graph.get(min_vertex, {}).items():
                if neighbour in visited:
                    continue
                new_distance = unvisited[min_vertex] + self.graph[min_vertex].get(neighbour, float("inf"))
                if new_distance < unvisited[neighbour]:
                    unvisited[neighbour] = new_distance
                    parents[neighbour] = min_vertex
            visited[min_vertex] = unvisited[min_vertex]
            unvisited.pop(min_vertex)
            if min_vertex == end:
                break
        return parents, visited

    @staticmethod
    def generate_path(parents, start, end):
        path = [end]
        while True:
            key = parents[path[0]]
            path.insert(0, key)
            if key == start:
                break
        return path
```

```python
input_graph=mat_dict                  #defined dictionary format for graph
input_vertices=[i for i in range(l)]  #defined vertices
#start_vertex = np.random.randint(0,l)
#end_vertex= np.random.randint(0,l)
start_vertex=22                       #starting point of path
end_vertex=4                          #ending point of path
dijkstra = Dijkstra(input_vertices, input_graph)
p, v = dijkstra.find_route(start_vertex, end_vertex)
print("Distance from %s to %s is: %.2f" % (start_vertex, end_vertex, v[end_vertex]))
se = dijkstra.generate_path(p, start_vertex, end_vertex)
print(se)                             #Displaying Nodes of optimized path
```

```python
points=[node[s] for s in se]               #Storing points for travelling
points=[[n-1-r,c] for r,c in points]       #Inverting vertical (row) value
print(points)
px=[]
for idx in range(len(se)-1):
    steps=np.array(points[idx+1])-np.array(points[idx])
    index=[0 if steps[0]!=0 else 1]                              #0 for x value linspace, 1 for y value linspace
    step=int(''.join([str(s) for s in steps[index]]))            #converting to integer
    p=np.linspace(points[idx], points[idx+1], abs(step)+1)
    p=p.tolist()
    px=px+p

px=[[x,y] for [y,x] in px]            #Exchanging x and y values from row and column values
# Here origin is mapped to bottom left corner for easier plotting input to matplotlib
```
